# Image results

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# test = ['cameraman', 'lena', 'livingroom', 'mandril', 'pirate', 'woman_blonde', 'woman_darkhair']
test = ['anpage', 'bplyce', 'drbost', 'ksunth', 'martin', 'pmives', 'rnpwil', 'sbains', 'swewin', 'yfhsie']
cs = ['paillier', 'dp']
op = ['negate', 'logtransform','pwrtransform']
```

```python
plt.style.use('fivethirtyeight')
```

```python
fig = plt.figure()
```

```python
def show_results(c, o):
    for testimg in test:
        f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
        f.suptitle(f'{testimg}', size=16)
        ax1.set_title('original')
        ax2.set_title('reference')
        ax3.set_title('decrypted')
        ax1.imshow(Image.open(f'{c}_{o}_{testimg}_original.png'))
        ax2.imshow(Image.open(f'{c}_{o}_{testimg}_reference.png'))
        ax3.imshow(Image.open(f'{c}_{o}_{testimg}_decrypted.png'))
```

```python
show_results('dp', 'logtransform')
```

```python
show_results('paillier', 'logtransform')
```

```python
show_results('dp', 'negate')
```
